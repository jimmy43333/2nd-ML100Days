### Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,2,3,4])
plt.xlabel("x label")
plt.ylabel("y label")
plt.plot([1,2,3,4],[5,6,7,8],'ro')
plt.show()
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,'r--',t,t*5,'bs',t,t**3,'g^')
```

```python
x = np.arange(0,100,10)
y = np.random.rand(10)
plt.bar(x,y)
x2 = np.arange(0,100,1)
y2 = np.random.rand(100)
plt.scatter(x2,y2)
```

```python
Tree = np.arange(0,500,1)
High = np.round(np.random.uniform(1,20,500),2)
Width = np.round(np.random.uniform(1,10,500),2)
Country = np.random.randint(0,3,500)
data_labels = ["TreeIndex","High(m)","Width(m)","Country"]
data_cols = [Tree,High,Width,Country]
dataframe = pd.DataFrame(dict(zip(data_labels,data_cols)))

color = {0:'r',1:'g',2:'b'}
dataframe.insert(4,column="color",value=dataframe["Country"].map(color))
dataframe.head(5)

dataframe.plot.scatter(x="High(m)",y="Width(m)",c=dataframe["color"])
dataframe.plot.bar(x="High(m)",y="Width(m)", rot=0)
```

### Seaborn

```python
import seaborn as sns
```

```python
Tree = np.arange(0,100,1)
High = np.round(np.random.uniform(1,20,100),2)
Width = np.round(np.random.uniform(1,10,100),2)
Country = np.random.randint(0,3,100)
data_labels = ["TreeIndex","High(m)","Width(m)","Country"]
data_cols = [Tree,High,Width,Country]
dataframe = pd.DataFrame(dict(zip(data_labels,data_cols)))

color = {0:'r',1:'g',2:'b'}
dataframe.insert(4,column="color",value=dataframe["Country"].map(color))
CountryName = {0:'Taiwan',1:'Japan',2:'China'}
dataframe["N"] = dataframe["Country"].map(CountryName)
sns.lmplot('High(m)','Width(m)',data=dataframe,fit_reg=False,hue='N')
dataframe.head(5)
```

```python
sns.boxplot(x=dataframe['High(m)'])
```
