## NOTE: How to use Pandas 
main data types : Series (處理時間序列,主要為一維陣列), DataFrame（處理結構化資料,如關聯式資料庫csv等等）
#### Read CSV file and HTML
```
df = pd.read_csv("filename",encoding="ISO-8859-1/utf-8")
ds = pd.read_html("http://...//")
```
#### Create Series
the data structure can be array, dictionary, single data
```
select = pd.Series(data,index = idx)
arr = ["A","B","C","D"]
select = pd.Series(cars)
dict = {
    "aaa": "111",
    "bbb": "222",
    "ccc": "333"
} 
select = pd.Series(dict,index = dict.keys())
print(select[[0,2,3]])
```
#### Create and Operation on DataFrame
```
df = pd.DataFrame(arr,columns=["name","value"])
.shape      # 回傳列數與欄數 
.describe()   # 回傳描述性統計
.heads(n)     # 回傳前n筆觀測值
.tail(n)      # 回傳後n筆觀測值 
.columns      # 回傳欄位名稱  
.index        # 回傳 index 
.info       # 回傳資料內容  
```
資料選擇及新增
```
df["Name"]

.loc(row_label,col_label)

```
資料排序

.sort_index(axis = 0, ascending = True)
.sort_values(by = 'num')
